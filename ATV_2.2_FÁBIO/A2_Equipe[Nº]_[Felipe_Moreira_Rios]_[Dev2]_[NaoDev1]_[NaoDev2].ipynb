## üîß Pr√©-processamento (GDAL)

   - Camadas Light/Dark

```python
%%bash
# 1) Instala o GDAL
apt-get update && apt-get install -y gdal-bin

# 2) Defina os nomes base dos shapefiles
shapefiles=(
  APPS
  AREA_CONSOLIDADA
  AREA_IMOVEL
  AREA_POUSIO
  HIDROGRAFIA
  RESERVA_LEGAL
  SERVIDAO_ADMINISTRATIVA
  USO_RESTRITO
  VEGETACAO_NATIVA
)

# 3) Loop para corrigir cada shapefile
for shp in "${shapefiles[@]}"; do
  echo "üîÑ Corrigindo ${shp}_1.shp ‚Üí ${shp}_fixed.shp"
  ogr2ogr \
    -f "ESRI Shapefile" \
    -makevalid \
    -s_srs EPSG:4326 \
    -t_srs EPSG:4326 \
    -nlt PROMOTE_TO_MULTI \
    -skipfailures \
    "/content/${shp}_fixed.shp" \
    "/content/${shp}_1.shp"
done
```

```python
!pip install geopandas requests folium matplotlib certifi urllib3

import geopandas as gpd
import os
import folium
import matplotlib.pyplot as plt
from folium import plugins

# pasta onde est√£o os shapefiles corrigidos
folder_path = '/content'
# apenas as camadas essenciais
shapefiles = [
    "APPS",            # √Åreas de Preserva√ß√£o Permanente
    "VEGETACAO_NATIVA",# Cobertura Vegetal Nativa
    "RESERVA_LEGAL",   # √Årea de Reserva Legal
    "HIDROGRAFIA",     # Corpos d'√°gua
    "AREA_IMOVEL"      # Limites dos im√≥veis
]

# cores distintas para cada camada essencial
color_map = {
    "APPS": "#1f77b4",
    "VEGETACAO_NATIVA": "#17becf",
    "RESERVA_LEGAL": "#8c564b",
    "HIDROGRAFIA": "#9467bd",
    "AREA_IMOVEL": "#2ca02c"
}

# 1) cria o mapa base com duas op√ß√µes de tiles
m = folium.Map(location=[-8.0, -36.5], zoom_start=7, control_scale=True)
folium.TileLayer('cartodbpositron', name='Light').add_to(m)
folium.TileLayer('cartodbdark_matter', name='Dark').add_to(m)

for shp in shapefiles:
    # usa o shapefile j√° corrigido pelo ogr2ogr
    path = os.path.join(folder_path, f"{shp}_fixed.shp")
    gdf = gpd.read_file(path)

    # opcional: amostra 20% para acelerar (descomente se necess√°rio)
    gdf = gdf.sample(frac=0.4, random_state=1)

    # 2) reparar e simplificar geometrias
    gdf['geometry'] = gdf.geometry.buffer(0).simplify(0.01)

    # 3) criar um FeatureGroup para ligar/desligar
    fg = folium.FeatureGroup(name=shp, show=False)

    # 4) adicionar GeoJson estilizado
    folium.GeoJson(
        data=gdf,
        style_function=lambda feat, col=color_map[shp]: {
            'fillColor': col, 'color': col, 'weight': 1, 'fillOpacity': 0.4
        },
        highlight_function=lambda _: {'weight': 3, 'fillOpacity': 0.7},
        tooltip=folium.GeoJsonTooltip(fields=[], aliases=[], labels=False)
    ).add_to(fg)

    # 5) popups nos centroides
    for _, row in gdf.iterrows():
        cen = row.geometry.centroid
        popup_html = (
            f"<b>{shp.replace('_',' ').title()}</b><br>"
            f"√Årea ‚âà {row.geometry.area:.0f} m¬≤<br>"
            f"Centroide: {cen.y:.5f}, {cen.x:.5f}"
        )
        folium.Marker(
            [cen.y, cen.x],
            icon=folium.Icon(color='blue', icon='info-sign'),
            popup=folium.Popup(popup_html, max_width=250)
        ).add_to(fg)

    fg.add_to(m)

    # 6) gerar mapa est√°tico PNG
    fig, ax = plt.subplots(figsize=(6,6))
    gdf.plot(ax=ax, color=color_map[shp], edgecolor='black', alpha=0.5)
    ax.set_title(f"{shp.replace('_',' ').title()} (static)")
    ax.set_xlabel('Longitude')
    ax.set_ylabel('Latitude')
    plt.tight_layout()
    fig.savefig(f"/content/{shp}_static_map.png", dpi=150)
    plt.close(fig)

# 7) controles e plugins
folium.LayerControl(collapsed=False).add_to(m)
plugins.MiniMap(toggle_display=True).add_to(m)
plugins.MousePosition().add_to(m)
plugins.Fullscreen().add_to(m)

# 8) salvar HTML interativo
out_html = '/content/Mapa_CAR_PSA_interativo.html'
m.save(out_html)
print(f"‚úÖ Mapa interativo salvo em: {out_html}")
print("‚úÖ Mapas est√°ticos (.png) salvos ao lado dos shapefiles corrigidos.")
```
